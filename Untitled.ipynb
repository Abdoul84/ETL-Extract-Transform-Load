```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
query = pd.read_csv("Resources/query_results.csv")
query
#query['Obesity'] = query['Obesity'].astype(float)
query['Positive_Affect'] = query['Positive_Affect'].astype(float)
```

```python
query = query.sort_values(by = 'Positive_Affect', axis=0, ascending=True, inplace=False)
query.head(15)
```

```python
happiness = query['Positive_Affect']
obesity = query['Obesity']
plt.xticks(rotation =90)
plt.scatter(obesity, happiness, edgecolors='r')
plt.ylabel('Positive Affect')
plt.xlabel('Obesity %')
plt.title('Obesity in each Country')
plt.show()
```
